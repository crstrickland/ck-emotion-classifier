```python
import keras
from keras.preprocessing.image import ImageDataGenerator
from keras.models import load_model
from keras.callbacks import ModelCheckpoint

from __future__ import absolute_import
from __future__ import division
from __future__ import print_function

import tensorflow as tf
import os
import numpy as np
import cv2
import random
import matplotlib.pyplot as plt
import sklearn
from sklearn.metrics import confusion_matrix
import gc

tf.logging.set_verbosity(tf.logging.INFO)

%matplotlib inline

# emotion coding: 0=neutral, 1=anger, 2=contempt, 3=disgust, 4=fear, 5=happy, 6=sadness, 7=surprise
```

```python
train_dir = os.path.join("aligned", "train")
valid_dir = os.path.join("aligned", "valid")
batch_size = 10
nb_train = 931
nb_valid = 381
img_w, img_h = 96, 96
steps_per_epoch = nb_train // batch_size
valid_steps = nb_valid // batch_size
epochs = 100
check_path = 'models/openface/openface_nn4_class-{epoch:02d}-{val_acc:.2f}.hdf5'
checkpoint = ModelCheckpoint(check_path, monitor='val_acc', verbose=1, save_best_only=True, mode='max')
callbacks_list = [checkpoint]
```

```python
train_datagen = ImageDataGenerator(
                rescale = 1./255,
                horizontal_flip = True,
                data_format='channels_first')

test_datagen = ImageDataGenerator(rescale=1./255,
                                 data_format='channels_first')
```

```python
train_data = train_datagen.flow_from_directory(train_dir, 
                                               batch_size = batch_size, 
                                               target_size = (96, 96), 
                                               class_mode="categorical")
valid_data = test_datagen.flow_from_directory(valid_dir, 
                                              batch_size = batch_size, 
                                              target_size = (96, 96), 
                                              class_mode="categorical")
```

```python
images_to_show = 1
vis_datagen = ImageDataGenerator()
vis_data = vis_datagen.flow_from_directory(train_dir, batch_size = images_to_show, target_size=(96, 96), class_mode="categorical", save_to_dir='genimg')
x_batch, y_batch = next(vis_data)

for i in range(images_to_show):
    print(y_batch[i])
    image = x_batch[i]
    plt.imshow(image)
    plt.show()
    
    
```

```python
model = load_model('models/nn4.small2.v7.py36.h5')
model.layers.pop()

for layer in model.layers:
    layer.trainable = False

x = model.layers[-1].output

x = keras.layers.Dense(8, activation='softmax', kernel_initializer='normal')(x)
model = keras.models.Model(inputs=model.input, outputs = x)
model.compile(optimizer='adagrad', loss='categorical_crossentropy', metrics=['accuracy'])

print(model.summary())

```

```python
model.fit_generator(train_data, 
                    steps_per_epoch = steps_per_epoch, 
                    epochs = epochs, 
                    validation_data = valid_data, 
                    validation_steps = valid_steps,
                    callbacks=callbacks_list)
model.save('try_5.h5')
```

```python
nb_valid = 327
conf_batch_size = 6
conf_epochs = nb_valid // conf_batch_size
conf_dir = os.path.join("aligned", "test")

conf_datagen = ImageDataGenerator(rescale=1./255,
                                 data_format='channels_first')
conf_data = conf_datagen.flow_from_directory(conf_dir, 
                                           batch_size = conf_batch_size, 
                                           target_size=(96, 96), 
                                           class_mode="categorical")

c_pred = []
c_true = []

for e in range(conf_epochs):
    x_batch, y_batch = next(conf_data)
    c_true = np.append(c_true, np.argmax(y_batch, axis=1))
    preds = tmodel.predict(x_batch, batch_size=conf_batch_size)
    c_pred = np.append(c_pred, np.argmax(preds, axis=1))
    
print(len(c_pred), len(c_true))
```

```python
labels = ['anger', 'contempt', 'disgust', 'fear', 'happiness', 'sadness', 'surprise']
confusion = confusion_matrix(c_true, c_pred)

print(confusion)
```

```python
plt.imshow(confusion, interpolation='nearest', cmap=plt.cm.PuRd)
plt.colorbar()
plt.title('Confusion Matrix')
plt.xticks(np.arange(len(labels)), labels, rotation=45)
plt.yticks(np.arange(len(labels)), labels)
plt.tight_layout()
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
```

```python
scores = {}

for path, name, files in os.walk('Emotion'):

        for file in files:
            if(file):
                with open(os.path.join(path, file)) as curf:

                    score = int(curf.read().strip(' ')[:1])
                    scores[score] = scores.get(score, 0) + 1

print(scores)
```

```python
nb_eval = 327
eval_batch_size = 6
eval_epochs = nb_eval // eval_batch_size
eval_dir = os.path.join("aligned", "test")

eval_datagen = ImageDataGenerator(rescale=1./255,
                                 data_format='channels_first')
eval_data = eval_datagen.flow_from_directory(eval_dir, 
                                           batch_size = eval_batch_size, 
                                           target_size=(96, 96), 
                                           class_mode="categorical")
```

```python
models = [os.path.join('models', 'openface', model) for model in os.listdir('models/openface')]
model_results = []

for mpath in models:
    m = load_model(mpath)
    
    ev = m.evaluate_generator(eval_data, eval_epochs)
    
    zipped = zip(m.metrics_names, ev)
    model_results.append((list(zipped), mpath))
    del m
    del ev
    gc.collect()
    
```

```python
# we need some test data
# these are less 'prime' images, so performance on them may be worse

def move_test_images():
    test_path = os.path.join("aligned", "test")
    test = 0
    
    for path, name, files in os.walk('Emotion'):

        for file in files:
            if(file):
                with open(os.path.join(path, file)) as curf:
                    # we have already taken away four images from the main folder
                    n = int(file[-14:-12]) - 4
                    print(n)
                    
                    score = int(curf.read().strip(' ')[:1]) + 1
                    print("score", score)
                    
                    for i in range(1):
                        fi = file[:-14] + str(n - i).zfill(2) + ".png"
                        
                        test += 1
                        
                        print("moved", os.path.join("aligned", fi), "to", os.path.join(test_path, str(score), fi))
                        try:
                            os.rename(os.path.join("aligned", fi), os.path.join(test_path, str(score), fi))
                        except:
                            pass
                        
    print(test, "test images")
    
move_test_images()
```
